### Import the test sets (predicted/estimated value) and compute the RMSE with respect to the true value

##### Test set (normal loading)

```python
for i in range(1, 5):
    file = f"v_test_pred_{i}.csv"
    v_test_pred = pd.read_csv(file)
    v_test_pred = v_test_pred.loc[:, (v_test_pred.columns != 'scenario') & (v_test_pred.columns != 'v_1')]
    print("Overall RMSE of model {} in terms of V: {}".format(i, mean_squared_error(v_test, v_test_pred, squared = False)))
```

##### Test set (increase in load)

```python
for i in range(1, 5):
    file = f"v_test_load_increase_pred_{i}.csv"
    v_test_load_increase_pred = pd.read_csv(file)
    v_test_load_increase_pred = v_test_load_increase_pred.loc[:, (v_test_load_increase_pred.columns != 'scenario') & (v_test_load_increase_pred.columns != 'v_1')]
    print("Overall RMSE of model {} in terms of V: {}".format(i, mean_squared_error(v_test_load_increase, v_test_load_increase_pred, squared = False)))
```
