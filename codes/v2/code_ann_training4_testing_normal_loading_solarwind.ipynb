### Train ANN models using training group 4

##### Using training group 4

```python
model_4 = Sequential()
model_4.add(Dense(64, input_dim = 64, activation = 'relu'))
model_4.add(Dense(64, activation = 'relu'))
model_4.add(Dense(64, activation = 'relu'))
model_4.add(Dense(9))

model_4.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['mean_squared_error'])

early_stopping = EarlyStopping(monitor = 'val_loss', patience = 5000, restore_best_weights = True, verbose = 1)
```

```python
model_4.fit(x_train_4,
            y_train_4,
            validation_data = (x_val, y_val),
            batch_size = 96,
            epochs = 100000,
            callbacks = [early_stopping],
            verbose = 2)

model_4.save('model_4.h5')
```

### Test the trained models on the test set

```python
model_4 = load_model('model_4.h5')
```

##### Using model 4 that was trained using training group 4
##### Test set (normal loading)

```python
y_test_pred_4 = model_4.predict(x_test)
y_test_pred_4 = pd.DataFrame(data = y_test_pred_4,
                                 columns = ['pg_1', 'pg_18', 'v_18', 'pg_22', 'v_22', 'pg_25', 'v_25', 'pg_33', 'v_33'])
y_test_pred_4.head()
```
