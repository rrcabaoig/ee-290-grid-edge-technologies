### Separate the dependent variables (optimal generator setpoints $P_{G_{j}}$ and $V_{j}$ for all buses with generators $j\epsilon \Omega_{G}$) and the independent variables (non-dispatchable loads $P_{D_{i}}$ and $Q_{D_{i}}$)

```python
x_train_1 = pd_qd_train_1.loc[:, pd_qd_train_1.columns != 'scenario']
x_train_1.head()
```

```python
y_train_1 = pg_v_train_1.loc[:, pg_v_train_1.columns != 'scenario']
y_train_1.head()
```

```python
x_train_2 = pd_qd_train_2.loc[:, pd_qd_train_2.columns != 'scenario']
x_train_2.head()
```

```python
y_train_2 = pg_v_train_2.loc[:, pg_v_train_2.columns != 'scenario']
y_train_2.head()
```

```python
x_train_3 = pd_qd_train_3.loc[:, pd_qd_train_3.columns != 'scenario']
x_train_3.head()
```

```python
y_train_3 = pg_v_train_3.loc[:, pg_v_train_3.columns != 'scenario']
y_train_3.head()
```

### Train ANN models using the 3 different training groups

```python
model_1 = Sequential()
model_1.add(Dense(64, input_dim = 64, activation = 'relu'))
model_1.add(Dense(64, activation = 'relu'))
model_1.add(Dense(64, activation = 'relu'))
model_1.add(Dense(10))
```

```python
model_1.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['mean_squared_error'])
```

```python
model_1.summary()
```

```python
model_1.fit(x_train_1, y_train_1, batch_size = 96, epochs = 5000)
```

```python
y_train_1_predicted = model_1.predict(x_train_1)
y_train_1_predicted_pd = pd.DataFrame(data = y_train_1_predicted,
                                     columns = ['pg_1', 'v_1', 'pg_18', 'v_18', 'pg_22', 'v_22', 'pg_25', 'v_25', 'pg_33', 'v_33'])
y_train_1_predicted_pd.head()
```
